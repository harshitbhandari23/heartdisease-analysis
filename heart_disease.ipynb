# ananlyzing numerical columns

# distribution of age

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
snp.distplot(x=df["Age"])
plt.subplot(2,2,2)
snp.distplot(x=df["RestingBP"],color="red")
plt.subplot(2,2,3)
snp.distplot(x=df["Cholesterol"],color="purple")
plt.subplot(2,2,4)
snp.distplot(x=df["MaxHR"])
```

# pie chart for chestpain

```python
a=df["ChestPainType"].value_counts().reset_index()
plt.figure(figsize=(10,5))
plt.title("chest pain type")
plt.pie(a["count"],labels=a["ChestPainType"],autopct="%0.1f%%",wedgeprops={"edgecolor":"black"})
```

# pie chart for stslope

```python
a=df["ST_Slope"].value_counts().reset_index()
plt.figure(figsize=(10,5))
plt.pie(a["count"],labels=a["ST_Slope"],wedgeprops={"edgecolor":"black"},autopct="%0.1f")
```

# piechart for resting ecg

```python
a=df["RestingECG"].value_counts().reset_index()
plt.figure(figsize=(10,5))
plt.pie(a["count"],labels=a["RestingECG"],autopct="%0.1f",wedgeprops={"edgecolor":"black"})
```
